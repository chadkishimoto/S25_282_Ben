**Exercise:** Make a plot of the displacement vs. time of a particle undergoing simple harmonic motion with damping, where $$x(t) = A e^{- b t} \cos( \omega t + \phi ),$$ using $A = 3~{\rm cm}$, $b = 0.3~{\rm s}^{-1}$, $\omega = 1.5~{\rm rad/s}$, $\phi = \pi / 3$.  (Note that $\pi$ can be represented as ``np.pi`` and the cosine function is ``np.cos( )``.)  

Make sure that your variables are approrpriately and descriptively named. Finally, make sure that your graph looks nice, and is well labeled.

```python
def displace(t):
    A = 3
    W = 1.5
    P = np.pi
    b = 0.3
    return A * (np.exp(-b * t)) * np.cos((W * t) + P)
time = np.linspace(-1,16,1000)
displacement = np.zeros(len(time))
for i in range(len(time)):
    displacement[i] = displace(time[i])
plt.figure()
plt.plot(time,displacement)
plt.xlabel("time")
plt.ylabel("displacement (in cm)")

plt.show()
```

**Exercise:** Create a parametric plot, where the points $(x, y) = (\cos \theta,\,\sin \theta)$ for $0 \leq \theta \leq 2\pi$.

```python
def angle(t):
    y = np.sin(t)
    x = np.cos(t)
    return x , y
    
theta = np.linspace(0,2*np.pi,1000)
x = np.zeros(len(theta))
y = np.zeros(len(theta))
for i in range(len(theta)):
    x[i] , y[i] = angle(theta[i])



plt.figure(figsize = (5,5))
plt.ylabel("sin theta")
plt.xlabel("cos theta")
plt.plot(x,y)
plt.figure()
plt.show()
```

```python
time, meters, v = np.loadtxt('Lab2.csv', delimiter=',', usecols=(0,1,2), unpack=True)

m = 2


def energy(meters, v):
    return ((1/2) * m * (v**2)) + ( m * 9.8 * meters)

def momentum(v):
    return m * v

moment = (np.zeros(len(v)))
for i in range(len(v)):
    moment[i] = momentum(v[i])
    
energ = np.zeros(len(meters))
for i in range(len(meters)):
    energ[i] = energy(meters[i], v[i])



plt.figure(figsize = (5,5))

plt.xlabel("time")

plt.plot(time,moment, color='red', marker='+', linestyle='dashed', label="momentum vs time")

plt.plot(time,energ, color='purple', marker='x', linestyle='dashdot', label="energy vs time")
plt.legend(loc='upper right')
plt.figure()

plt.show()
```
