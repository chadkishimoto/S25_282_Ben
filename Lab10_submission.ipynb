```python
import numpy as np
import matplotlib.pyplot as plt
```

Initialize a random number generator, then 
- create an array of 1,000,000 standard normal random numbers.
- create 100 samples, each with a slice of 10,000 of these numbers
- from each sample, determine the mean and standard deviation
- **create a histogram** of the means
- **create a histogram** of the standard deviations
- **print** the mean and standard deviation of the mean
- **print** the mean and standard deviation of the standard deviation

```python
rng = np.random.default_rng()
z = rng.standard_normal(1000000)
test_array = np.arange(len(z))
mean_array = np.zeros(100)
std_array = np.zeros(100)
for i in range(100):
    mean_array[i] = np.mean(z[i*10000 :(i+1)*10000 ])
    std_array[i] = np.std(z[i*10000 :(i+1)*10000 ])
plt.figure()
plt.hist(mean_array, 100)
plt.show()
plt.figure()
plt.hist(std_array, 100)
plt.show()
print(f" Mean for mean_array is {np.mean(mean_array)} and Standard deviation for mean_array is {np.std(mean_array)}")
print(f" Mean for std_array is {np.mean(std_array)} and Standard deviation for std_array is {np.std(std_array)}")
```

Create 2000 two-dimensional random walks. Then create a plot of the last 100 of these walks (y vs. x).

```python
def gaussian_random_walk(N):
    walk = np.zeros(N)
    for i in range(1, N):
        walk[i] = walk[i-1] + rng.standard_normal()
    return walk

time = np.arange(100)

matrix1 = np.zeros((2000,100))
matrix2 = np.zeros((2000,100))

for i in range(2000):
        matrix1[i] = gaussian_random_walk(100)
        matrix2[i] = gaussian_random_walk(100)


plt.figure()   # First 100 walks. 
for i in range(100):
    plt.plot(matrix1[i,:], matrix2[i,:])
plt.title("First 100 walks")    
plt.show()

plt.figure() # Last 100 walks.
for i in range(1,101):
    plt.plot(matrix1[-i,:], matrix2[-i,:])
plt.title("Last 100 walks")
plt.show()
```

With the two-dimensional random walks. Given that $r = \sqrt{x^2 + y^2}$, find the final $r$ value for all 2000 walks and
- Create a histogram of these r values.
- Print the mean and standard deviation of these r values

```python
r_matrix = np.zeros((2000,100))
r_matrix = np.sqrt(matrix1**2 + matrix2**2)

for i in range(100):
    plt.plot(time, r_matrix[i,:])
plt.title("r values vs time")
plt.show()

plt.hist(r_matrix[:,-1])
plt.title("r values")
plt.show()


r_mean = np.zeros(20)
r_std = np.zeros(20)

for i in range(20):
    r_mean[i] = np.mean(r_matrix[i*100 :(i+1)*100 ])
    r_std[i] = np.std(r_matrix[i*100 :(i+1)*100 ])


print(f" The best value for r is: {np.mean(r_mean)} and the uncertianty is: {np.std(r_mean)}")
```
