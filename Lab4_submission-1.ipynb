**Exercise:** Create a heat map plot of the function
$$ f(x, y) = e^{- (x^2 + y^2)}$$
Using:
- 50 points in x from -2 to +2
- 200 points in y from -4 to +4

```python
x_values = np.linspace(-2,2,100)   ## This creates an array of x values ##
y_values = np.linspace(-2,2,100)   ## This creates an array of y values ##

def f_of_values(y_value, x_value):                  ## function  for z values##
    return np.e**(-(((x_value)**2)+(y_value)**2))  

M = np.zeros((len(y_values),len(x_values)))  ## Array for z values ##

for i in range(len(y_values)):  ## nested for loop to create all possible z values in the arrays created ##
    for j in range(len(x_values)):
        M[i,j] = f_of_values(y_values[i],x_values[j])

plt.figure(figsize=(8,8))
plt.imshow(M,extent=[-2,2,-4,4]) 
plt.colorbar()
plt.show()
```

**Exercise:**  Plot a heat map of the function
$$ g(x, y) = e^{- x^2} e^{- y^2 / 4}$$
using the same x and y arrays.

```python
x_values = np.linspace(-2,2,50)
y_values = np.linspace(-4,4,200)

G = np.zeros((len(y_values),len(x_values)))

def g_of_values(y_value, x_value):
    return np.e**(-((x_value)**2))*(np.e**(-(((y_value)**2)/4)))


for i in range(len(y_values)):
    for j in range(len(x_values)):
        G[i,j] = g_of_values(y_values[i],x_values[j])

plt.figure(figsize=(8,8))
plt.imshow(G,extent=[-2,2,-4,4]) 
plt.colorbar()
plt.show()
```
